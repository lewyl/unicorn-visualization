```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./Unicorn+Companies/Unicorn_Companies.csv")
```

```python
df.head()
```

```python
def convert_value(val_in_str):
    unit = val_in_str[-1]
    if unit == "B":
        val_in_int = int(val_in_str[1:len(val_in_str)-1])*10**9
    elif unit == "M":
        val_in_int = int(val_in_str[1:len(val_in_str)-1])*10**6
    else:
        val_in_int = np.NaN
    return val_in_int
```

```python
df['Valuation'] = df['Valuation'].apply(convert_value)
df['Funding'] = df['Funding'].apply(convert_value)
```

```python
df.head()
```

```python
def merge_industry(industry):
    if industry == "Artificial Intelligence":
        return "Artificial intelligence"
    else:
        return industry

df['Industry'] = df['Industry'].apply(merge_industry)
```
